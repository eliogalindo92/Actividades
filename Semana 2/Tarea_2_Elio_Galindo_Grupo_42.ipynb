```python
# Tarea 2. Elio Galindo Fonseca. Grupo 42

!git clone https://github.com/eliogalindo92/Actividades.git
import cv2
import random
import numpy as np
from google.colab.patches import cv2_imshow
from IPython.display import clear_output
clear_output()

# Cargo la imagen de disco
image = cv2.imread('/content/opencv_resources/Semana 2/Recursos/land.jpg')

# Inciso a

# Roto la imagen 90° en el sentido de las manecillas
rows, cols, channel = image.shape
matrix = cv2.getRotationMatrix2D((cols/2, rows/2), -90, 1)
rotated_image = cv2.warpAffine(image, matrix, (cols, rows))

# Utilizo la función hstack de numpy para mostrar las imagenes en horizontal
horizontal_stack = np.hstack((image, rotated_image))
cv2_imshow(horizontal_stack)
```

```python
# Inciso b

# Creo los kernels
kernel_1 = np.ones((5, 5), np.float32)/25
kernel_2 = np.ones((9, 9), np.float32)/25

# Creo un desenfoque personalizado con los kernels
blurred_image_1 = cv2.filter2D(image, -1, kernel_1)
blurred_image_2 = cv2.filter2D(image, -1, kernel_2)

# Utilizo la función hstack de numpy para mostrar las imagenes en horizontal
horizontal_stack = np.hstack((image, blurred_image_1, blurred_image_2))
cv2_imshow(horizontal_stack)
```

```python
# Inciso C

# Función para agregar ruido de sal y pimienta
def salt_and_pepper(image, percentage):
    value = int(percentage * image.shape[0] * image.shape[1]) # Número de puntos de ruido de sal y pimienta
    random.randint(0, image.shape[0])
    copied_image= image.copy()
    for i in range(value):
        x = random.randint(0, copied_image.shape[0]-1) # Un número entero aleatorio desde 0 hasta la longitud de la imagen, porque es un intervalo cerrado, -1
        y = random.randint(0, copied_image.shape[1]-1)
        if random.randint(0, 1) == 0: # Probabilidad en blanco y negro
            copied_image[x, y] = (255, 255, 255) # ruido blanco
        else:
            copied_image[x, y] =(0, 0, 0) # ruido negro
    return copied_image

noised_image = salt_and_pepper(image, 0.05) # 5 por ciento de ruido de sal y pimienta

# Utilizo la función hstack de numpy para mostrar las imagenes en horizontal
horizontal_stack = np.hstack((image, noised_image))
cv2_imshow(horizontal_stack)
```

```python
# Inciso d

# Elimino el ruido de sal erosionando la imagen para hacer más pequeños los objetos
# en primer plano y agrandar los agujeros dentro de la imagen

# Creo un kernel 
kernel_3 = np.ones((3, 3), dtype="uint8") 
eroded_image = cv2.erode(noised_image, kernel_3, 1)

# Utilizo la función hstack de numpy para mostrar las imagenes en horizontal
horizontal_stack = np.hstack((noised_image, eroded_image))
cv2_imshow(horizontal_stack)
```
