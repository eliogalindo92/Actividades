```python
#--- Tarea 1. Elio Galindo Fonseca. Grupo 42 ---#

import cv2
from google.colab.patches import cv2_imshow
from skimage import io

# Leer la imagen
img_src = 'http://wx2.sinaimg.cn/mw690/ac38503ely1fesz8m0ov6j20qo140dix.jpg'
image = io.imread(img_src)
rgb_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Coordenadas del rectangulo
xmin, ymin = (50, 200)
xmax, ymax = (600, 650)

# Color, Grosor y Tipo de linea del rect√°ngulo
color = (0, 0, 255)
thickness = 3
linetype = cv2.LINE_AA

# Color, Grosor y Tipo de fuente
text_color = (0, 255, 0)
text_thickness = 2
position = (10, 150)
font = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 1
text = "Tarea 1. Elio Galindo Fonseca. Grupo 42"

# Dibujar texto
rgb_image = cv2.putText(rgb_image, text, position, font, font_scale, text_color, text_thickness)

# Dibujar rectangulo
rgb_image = cv2.rectangle(rgb_image, (xmin, ymin), (xmax, ymax), color, thickness, linetype)

# Mostrar imagen
cv2_imshow(rgb_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
